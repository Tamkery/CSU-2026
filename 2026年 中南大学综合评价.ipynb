# 中南大学2026年综合评价录取（云南省） 招生简章
发布时间：2026-06-18 14:11

## 概述申明

本人中南大学毕业，当年普通高考录取，到学校才知道面向云南和湖南省还有综合评价的渠道

打破信息孤岛,让有缘人能有更好的选择机会

本科普通类是双向选择的过程,综合评价个人选择的机会更大

---

## 一 什么是中南大学的综合评价录取方式?__云南省

全面贯彻落实《教育强国建设规划纲要（2024—2035年）》战略部署，执行《教育部办公厅关于做好2026年普通高等学校部分特殊类型招生工作的通知》（教学厅〔2025〕7号）及相关文件精神，探索**多维度考核评价模式**，学校2026年继续在**云南和湖南**省开展综合评价录取招生工作

走**提前本科批C段**，不占用本科普通类通道,多一次志愿机会

中南大学既是教育部直属高校，也是**中央直管副部级**建制高校。
‌教育部直属‌：学校由中华人民共和国教育部直接管理，是**首批国家“211工程”、“985工程”重点建设高校**，以及世界一流大学A类建设高校 。
‌中央直管‌：自2003年起，中南大学被明确为中央直管的副部级大学，其党委书记和校长由中共中央直接任命，属于全国31所中管高校之一 。

## 二 政策解读

每位考生须参加6个站点的考核，每个站点的考核时间为5分钟。
考核评价采用等级制，设A+、A、A-、B+、B、B-、C、D八个等级，A+的权重分值为19分，A的权重分值为18分，A-的权重分值为17分，B+的权重分值为16分，B的权重分值为15分，B-的权重分值为14分，C的权重分值为12分，D的权重分值为10分。各站点考核评价等级折算分值之和为面试考核成绩。

**时间**:7月6日,考生需携带本人**身份证原件**进行现场报到，并领取准考证

**地点**:云南大学东陆校区（昆明市五华区翠湖北路2号）

**费用**:参加综合评价录取考核的考生需缴纳考试费200元/人

**面试考核成绩**与**高考成绩**之和构成**综合评价成绩**。

站点包括**公民素质、科学思维、人文素养、人际交往、创新想象、个性特长**六个。每位考生须参加**6个站点**的考核，每个站点的考核时间为**5分钟**。

面试考核成绩不低于70分；(六个站点面试,C等级以上就有机会)
